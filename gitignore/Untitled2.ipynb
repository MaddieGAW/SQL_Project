```python
import pandas as pd
import requests

# Function to fetch ACLED data for a specific country and year
def fetch_acled_data(country, year):
    base_url = "https://api.acleddata.com/acled/read"
    params = {
        "email": "madison@madisonwells.co.uk",
        "key": "Fo3IDO3l9j4sKbAFUZBy",
        "terms": "accept",
        "iso": country,
        "year": year,
        "limit": 500,  
        "page": 1
    }
    
    all_data = []

    while True:
        response = requests.get(base_url, params=params)
        data = response.json()
        
        if not data.get('success', False) or 'data' not in data:
            break
        
        all_data.extend(data['data'])
        params['page'] += 1
        
        if len(data['data']) < params['limit']:
            break

    df = pd.DataFrame(all_data)
    
    if not df.empty:
        if 'country' in df.columns:
            df_country = df[df['country'] == country]
        else:
            df_country = pd.DataFrame()
        
        total_fatalities_in_country = df['fatalities'].astype(int).sum() if 'fatalities' in df else 0
    else:
        total_fatalities_in_country = 0

    return total_fatalities_in_country

# Apply the function to each row of the DataFrame
def calculate_country_fatalities(row, fetched_data):
    iso = row['ISO']
    year = row['YEAR']
    
    # Check if we already fetched data for this country and year
    if (iso, year) in fetched_data:
        return fetched_data[(iso, year)]
    else:
        # Fetch data and store it in fetched_data to avoid multiple calls for the same country and year
        fatalities = fetch_acled_data(iso, year)
        fetched_data[(iso, year)] = fatalities
        return fatalities

# Main function to calculate country fatalities for each row in the DataFrame
def calculate_country_fatalities_for_df(df):
    fetched_data = {}  # Dictionary to store fetched data for each country and year
    df['country_fatalities'] = df.apply(lambda row: calculate_country_fatalities(row, fetched_data), axis=1)
    return df
```
