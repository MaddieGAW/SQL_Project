```python
import pandas as pd
# Set the display options to show all columns
pd.set_option('display.max_columns', None)

# Replace 'file_path.xlsx' with the actual path to your Excel file
file_path1 = 'number_of_political_violence_events_by_country-year_as-of-12Apr2024.xlsx'

# Read Excel file into a DataFrame
df1 = pd.read_excel(file_path1)

# Replace 'file_path.xlsx' with the actual path to your Excel file
file_path2 = 'number_of_reported_fatalities_by_country-year_as-of-12Apr2024.xlsx'

# Read Excel file into a DataFrame
df2 = pd.read_excel(file_path2)
```

```python
df1
```

```python
df2
```

```python
# Merge the two DataFrames based on common columns 'Country' and 'Year'
merged_df = pd.merge(df1, df2, on=['Country', 'Year'])

# Display the merged DataFrame
merged_df.head()
```

```python
# Group the DataFrame by 'Year' and sum the 'Events' and 'Fatalities' columns
global_stats = merged_df.groupby('Year')[['Events', 'Fatalities']].sum().reset_index()

# Display the global statistics
global_stats
```

```python
import pandas as pd

# Replace 'file_path.csv' with the actual path to your CSV file
file_path = 'security_incidents_2024-04-17.csv'

# Read CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the DataFrame
df.head()  # Use df.head(n) to display first n rows
```

```python
# Drop the first row from the DataFrame
df = df.drop(0)

# Reset the index after dropping the row
df = df.reset_index(drop=True)

# Group the DataFrame by 'Year' and sum the 'Fatalities' column
deaths_per_year = df.groupby('Year')['Total killed'].sum().reset_index()

# Display the deaths per year
deaths_per_year
```

```python
import pandas as pd

# Assuming df is your DataFrame
# Convert 'Total killed' column to numeric
df['Total killed'] = df['Total killed'].astype(int)

# Group the DataFrame by 'Year' and sum the 'Total killed' column
total_killed_per_year = df.groupby('Year')['Total killed'].sum().reset_index()

# Display the total killed per year
print(total_killed_per_year)
```

```python
# Convert the 'Year' column in df to int
df['Year'] = df['Year'].astype(int)

# Merge the two DataFrames based on common column 'Year'
merged_df2 = pd.merge(merged_df, total_killed_per_year, on='Year')

# Display the merged DataFrame
merged_df2
```
